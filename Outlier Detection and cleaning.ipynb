# Removing Outliers

*Using Z-score >3 as outliers*

*Cheking Whether RestingBp,Cholesterol, FastingBS, MAxHR,OLdPeak has outlier*

```python
heart.shape
```

```python
heart[heart.RestingBP>heart.RestingBP.mean()+3*heart.RestingBP.std()]
```

```python
df1=heart[heart.RestingBP<=heart.RestingBP.mean()+3*heart.RestingBP.std()]
```

```python
df1.shape
```

```python
heart[heart.Cholesterol	>heart.Cholesterol	.mean()+3*heart.Cholesterol	.std()]
```

```python
df2=df1[df1.Cholesterol<=df1.Cholesterol.mean()+3*df1.Cholesterol.std()]
```

```python
df2.shape
```

```python
heart[heart.FastingBS>heart.FastingBS.mean()+3*heart.FastingBS.std()]
```

```python
heart[heart.MaxHR>heart.MaxHR.mean()+3*heart.MaxHR.std()]
```

```python
heart[heart.Oldpeak>heart.Oldpeak.mean()+3*heart.Oldpeak.std()]
```

```python
df3=df2[df2.Oldpeak<=df2.Oldpeak.mean()+3*df2.Oldpeak.std()]
```

```python
df3.shape
```

```python
df3.ChestPainType.unique()
```

```python
df3.RestingECG.unique()
```

```python
df3.ExerciseAngina.unique()
```

```python
df3.ST_Slope.unique()
```

```python
df4=df3.copy()
df4.Sex.replace({
    'M':1,
    'F':0
},inplace=True)

df4.ChestPainType.replace({
    'ATA':1,
    'NAP':2,
    'ASY':3,
    'TA':4
},inplace=True)

df4.RestingECG.replace({
    'Normal':1,
    'ST':2,
    'LVH':3
},inplace=True)

df4.ExerciseAngina.replace({
    'N':1,
    'Y':0
},inplace=True)

df4.ST_Slope.replace({
    'Up':1,
    'Flat':2,
    'Down':3
},inplace=True)
```

```python
df4.head(1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
inputs=df4.drop('HeartDisease',axis='columns')
output=df4.HeartDisease
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model1=DecisionTreeClassifier()
```

```python
X_train,X_test,y_train,y_test=train_test_split(inputs,output,test_size=0.2,random_state=20)
```

```python
model1.fit(X_train,y_train)
```

```python
model1.score(X_test,y_test)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scalar=StandardScaler()
```

```python
scaled_inputs=scalar.fit_transform(inputs)
```

```python
from sklearn.decomposition import PCA
```

```python
scaled_inputs.shape
```

```python
pca=PCA(0.95)
```

```python
pca_inputs.shape
```

```python
pca.explained_variance_ratio_
```

```python
pca.n_components_
```

```python
X_trainpca,X_testpca,y_trainpca,y_testpca=train_test_split(pca_inputs,output,test_size=0.2,random_state=20)
```

```python
model1.fit(X_trainpca,y_trainpca)
```

```python
model1.score(X_testpca,y_testpca)
```
