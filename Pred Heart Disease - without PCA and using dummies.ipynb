# Predicting if a person has heart disease Without Principle component analysis

```python
import pandas as pd
```

```python
heart=pd.read_csv('heart.csv')
```

```python
heart.head(1)
```

```python
heart.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.svm import SVC
```

```python
from sklearn.decomposition import PCA
```

```python
heart_labled=pd.get_dummies(heart,columns=['Sex','ChestPainType','RestingECG','ST_Slope','ExerciseAngina'],drop_first=True).astype(int)
```

```python
heart_labled.head(1)
```

```python
heart_labled.columns
```

```python
inputs=heart_labled.drop('HeartDisease',axis='columns')
output=heart_labled.HeartDisease
```

```python
X_train,X_test,y_train,y_test=train_test_split(inputs,output,test_size=0.2,random_state=20)
```

```python
model1=LogisticRegression(max_iter=70)
```

```python
model1.fit(X_train,y_train)
```

```python
model1.score(X_test,y_test)
```

# Random Forest

```python
model2=RandomForestClassifier(n_estimators=100)
```

```python
model2.fit(X_train,y_train)
```

```python
model2.score(X_test,y_test)
```

# Decision Tree

```python
model3=DecisionTreeClassifier()
```

```python
model3.fit(X_train,y_train)
```

```python
model3.score(X_test,y_test)
```

# Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
```

```python
model4=GaussianNB()
```

```python
model4.fit(X_train,y_train)
```

```python
model4.score(X_test,y_test)
```

# SVC

```python
model5=SVC(C=1,kernel='linear')
```

```python
model5.fit(X_train,y_train)
```

```python
model5.score(X_test,y_test)
```
