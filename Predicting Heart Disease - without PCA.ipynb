# Without Principle component analysis using Label Encocer

```python
import pandas as pd
```

```python
heart=pd.read_csv('heart.csv')
```

```python
heart.head(1)
```

```python
heart.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
Sex_le=le.fit_transform(heart.Sex)	
```

```python
ChestPainType_le=le.fit_transform(heart.ChestPainType)
```

```python
RestingECG_le=le.fit_transform(heart.RestingECG)
```

```python
ExerciseAngina_le=le.fit_transform(heart.ExerciseAngina)
```

```python
ST_Slope_le=le.fit_transform(heart.ST_Slope)
```

```python
heart['Sex_le']=Sex_le
heart['ChestPainType_le']=ChestPainType_le
heart['RestingECG_le']=RestingECG_le
heart['ExerciseAngina_le']=ExerciseAngina_le
heart['ST_Slope_le']=ST_Slope_le
```

```python
heart.head(1)
```

```python
heart.drop(['Sex','ChestPainType','RestingECG','ExerciseAngina','ST_Slope'],axis='columns',inplace=True)
```

```python
heart.head(1)
```

```python
heart_inputs=heart.drop('HeartDisease',axis='columns')
heart_output=heart.HeartDisease
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.svm import SVC
```

```python
from sklearn.decomposition import PCA
```

```python
X_train,X_test,y_train,y_test=train_test_split(heart_inputs,heart_output,test_size=0.2,random_state=20)
```

```python
model1=LogisticRegression(max_iter=1000)
```

```python
model1.fit(X_train,y_train)
```

```python
model1.score(X_test,y_test)
```

```python
from sklearn.metrics import classification_report
```

```python
log_predict=model1.predict(X_test)
```

```python
print(classification_report(y_test,log_predict))
```
