# Using Principle Component Analysis

```python
import pandas as pd
```

```python
heart=pd.read_csv('heart.csv')
```

```python
heart.head(1)
```

```python
heart.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
Sex_le=le.fit_transform(heart.Sex)	
```

```python
ChestPainType_le=le.fit_transform(heart.ChestPainType)
```

```python
RestingECG_le=le.fit_transform(heart.RestingECG)
```

```python
ExerciseAngina_le=le.fit_transform(heart.ExerciseAngina)
```

```python
ST_Slope_le=le.fit_transform(heart.ST_Slope)
```

```python
heart['Sex_le']=Sex_le
heart['ChestPainType_le']=ChestPainType_le
heart['RestingECG_le']=RestingECG_le
heart['ExerciseAngina_le']=ExerciseAngina_le
heart['ST_Slope_le']=ST_Slope_le
```

```python
heart.head(1)
```

```python
heart.drop(['Sex','ChestPainType','RestingECG','ExerciseAngina','ST_Slope'],axis='columns',inplace=True)
```

```python
heart.head(1)
```

```python
heart_inputs=heart.drop('HeartDisease',axis='columns')
heart_output=heart.HeartDisease
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.svm import SVC
```

```python
from sklearn.decomposition import PCA
```

```python
heart.shape
```

```python
pca=PCA(n_components=4)
```

```python
inputs_pca=pca.fit_transform(heart_inputs)
```

```python
inputs=inputs_pca
output=heart.HeartDisease
```

```python
inputs_pca.shape
```

```python
X_train,X_test,y_train,y_test=train_test_split(inputs_pca,output,test_size=0.2,random_state=20)
```

```python
model1=LogisticRegression()
```

```python
model1.fit(X_train,y_train)
```

```python
model1.score(X_test,y_test)
```

# Random Forest

```python
model2=RandomForestClassifier()
```

```python
model2.fit(X_train,y_train)
```

```python
model2.score(X_test,y_test)
```

# Decision Tree?

```python
model3=DecisionTreeClassifier()
```

```python
model3.fit(X_train,y_train)
```

```python
model3.score(X_test,y_test)
```

# SVC


```python
model4=SVC()
```

```python
model4.fit(X_train,y_train)
```

```python
model4.score(X_test,y_test)
```
